```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris["data"], columns=iris["feature_names"])
df["target"] = iris["target"]
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.countplot(df["target"], palette="hls")
```

```python
sns.countplot(y=df["target"], palette="Blues")
```

```python
sns.scatterplot(x=df["sepal length (cm)"], y=df["sepal width (cm)"], hue="target", data=df)
```

```python
sns.palplot(sns.cubehelix_palette(8))
```

```python
sns.distplot(df["sepal length (cm)"])
```

```python
grid = sns.PairGrid(df)
grid.map(sns.scatterplot, hue="target", data=df)
```

```python
grid = sns.FacetGrid(df, col="target")
grid.map(sns.distplot, "sepal width (cm)", color="green")
```

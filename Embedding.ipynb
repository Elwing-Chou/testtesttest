```python
from keras.datasets.mnist import load_data
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
import pandas as pd
x_train_reshape = x_train.reshape(-1, 784)
pd.DataFrame(x_train_reshape)
```

```python
from sklearn import datasets
mnist = datasets.load_digits()
X = mnist.data
Y = mnist.target
```

```python
from sklearn import decomposition
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
pca = decomposition.PCA(n_components=3)
new_X = pca.fit_transform(X)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(new_X[:, 0], new_X[:, 1], new_X[:, 2], c=Y, cmap="spectral")
plt.show()
```

```python
import seaborn as sns

pca = decomposition.PCA(n_components=2)
new_X = pca.fit_transform(X)
plt.figure(figsize=(10, 10))
sns.scatterplot(new_X[:, 0], new_X[:, 1], hue=Y, palette=sns.color_palette("hls", 10))
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
plt.figure(figsize=(10, 10))


tsne.fit_transform(X)
sns.scatterplot(tsne.embedding_[:, 0], tsne.embedding_[:, 1], hue=Y, palette=sns.color_palette("hls", 10))
```
